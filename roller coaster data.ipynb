```python
#univariate analysis
ax=df['Year_intro'].value_counts()\
.head(5)\
.plot(kind='bar', title='Top years coaster was introduced',)
ax.set_xlabel('Year Introduced')
ax.set_ylabel('Number')
```

```python
c=df.query('Location!="Other"')\
.groupby('Location')['speed_mph']\
.agg(['mean','count'])\
.query('count>=10')\
.sort_values('mean')['mean']\
.plot(kind='barh',title='Average Coast speed by location')
c.set_xlabel('avg coaster speed')
plt.show()
```

##### Highest coaster speed in Busch Gardens Williamsburg and lowest in Alton towers

```python
c=df.query('Location!="Other"')\
.groupby('Location')['speed_mph']\
.agg(['mean','count'])\
.query('count>=10')\
.sort_values('mean')\
.plot(kind='barh',title='Average Coast speed by location')
c.set_xlabel('avg coaster speed')
plt.show()
```

##### The highest coaster speed is at Busch Gardens Williamsburg whereas the lowest average roller coaster speed is at Alton towers. The number of coasters is higher and almost same in Cedar Point and Kings Island.
